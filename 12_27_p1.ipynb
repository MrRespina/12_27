```python
# slicing도 가능함.
df1.loc[0:9]   # 명시적 인덱스를 slicing 할 때는 리스트,튜플의 슬라이싱과 다르게 +1 하여 사용하지 않는다.
df1.iloc[0:0+10] # 암시적 인덱스를 slicing 할 때는 똑같이 +1 해줘야함.

df1.loc[0:9][:]
df1.loc[0:9,:]

df1.iloc[0:10][:]
df1.iloc[0:10,:]     # 이런 식으로 슬라이싱을 사용할 수 있다. 명시적 : 정확한 인덱스 범위 , 암시적 : 리스트처럼 1:1+3 이런 식.

df1.loc[0:9][['country','continent','year']]
df1.loc[0:9,['country','continent','year']]

df1.iloc[0:10,0:3]    # 암시적 인덱스는 튜플,리스트처럼 +1 해줘야 정확한 값이 나오는 것을 인지할 것.
df1[['country','continent','year']][0:10]     # 암시적 인덱스로 사용된다.

df1.iloc[0:10,[0,1,2]]
df1[['country','continent','year']]

df1.loc[: , ['country','continent','year']]
df1.iloc[: , [0,1,2]]
df1.iloc[: , 0:3]

# 밑의 메소드들은 모두 같은 결과를 도출하는 방법이다. 명시적/암시적
df1['year'][2]
df1.loc[2]['year']
df1.loc[2, 'year']
df1.iloc[2,2]
df1.iloc[-1704+2,2]
```

```python
# range() 함수 사용하기.
# 모든 행에서 3열~5열까지
df1.iloc[:,3:5+1]
df1.iloc[:,range(3,6)]

# 열을 0부터 5까지중 한칸씩 띄어서
df1.iloc[:, 0:6:2]
df1.iloc[:,range(0,6,2)]

# 행을 0부터 20개를 가져오고 열을 0부터 5까지 가져올 때 행과 열을 한칸씩 띄어서!
df1.iloc[0:0+20,0:6:2]

# 열을 두칸 씩
df1.iloc[:,::3]

# 모든 행을 두칸 씩
df1.iloc[::3,:]
```

```python
import numpy as np

m = np.array([1,2,3,4])

print(m<3)
print(m[m<3])
print(m==3)
print(m[m==3])

# 팬시 인덱스
df1['year'] == 1952
df1[df1['year'] == 1952]
df1[df1['year'] == 1952]['lifeExp']
df1[df1['year'] == 1952][['lifeExp','gdpPercap']]

# 집계 연산

# 1952 년도 데이터의 개수
(df1['year'] == 1952).sum()

# 평균 수명
print(df1[df1['year'] == 1952]['lifeExp'].mean())

# 평균 수명이 제일  작은 것?
print(df1[df1['year'] == 1952]['lifeExp'].min())

# 평균 수명이 제일 큰 것
print(df1[df1['year'] == 1952]['lifeExp'].max())

# 평균 수명의 합계
print(df1[df1['year'] == 1952]['lifeExp'].sum())
```

```python
# 1952 년 평균 수명의 평균
print(df1[df1['year'] == 1952]['lifeExp'].mean())

# 각 년도 별 평균 수명
print(df1.groupby('year')['lifeExp'].mean())

# 각 년도 별 최소 수명
print(df1.groupby('year')['lifeExp'].min())

# 각 년도 별 평균 수명의 최소값에서 1967년의 값
print(df1.groupby('year')['lifeExp'].min()[1967])

# 각 년도 별 대륙의 평균 수명을 팬시 index
print(df1.groupby(['year','continent'])['lifeExp'].mean())

# 각 대륙의 나라는 몇 개?    nunique() 로 중복값 제거
print(df1.groupby('continent')['country'].nunique())
```
